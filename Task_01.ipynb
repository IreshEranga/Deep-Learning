```python
exp_arr=np.random.exponential(size=(100000,1))
uniform_arr = np.random.uniform(size=(100000,1))
normal_arr = np.random.normal(size=(100000,1))

plt.hist(exp_arr, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_arr, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_arr, density=True, bins=100, histtype="step", color="blue", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Z = X^2 + Y^2 Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Display first 5 observations
df.head()
```

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')   # Color by evolution stage
```

```python
# Boxplot
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

```python
# Preprocess DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# New boxplot using stats_df
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=stats_df)
```

```python
# Violin plot
plt.figure(figsize=(12,8)) # Set plot dimensions
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
import seaborn as sns
import pandas as pd


features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
subset = df[features]

# Pearson correlation
pearson_corr = subset.corr(method='pearson')
# Spearman correlation
spearman_corr = subset.corr(method='spearman')

# Heatmap with values displayed
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation")
plt.show()
```
